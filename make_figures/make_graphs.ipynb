```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
csv=np.genfromtxt ('./data/fall22-phys180.tsv',
                   skip_header=2,
                   delimiter=",")
```

```python
length=csv[:,0]
time=csv[:,1]

# this is the small-oscillation physics result (eg, Pendulums, Open Stax University Physics, section 16.4)
l_th=np.arange(min(length),max(length))
g=9.81
t_th=(2*np.pi/np.sqrt(g))*(1./10.)*np.sqrt(l_th)
```

```python

plt.plot(l_th,t_th,"-",markerfacecolor="none",label="$T=(2 \pi) \sqrt{L/g}$")

plt.plot(length,time,"o",markerfacecolor="none", label="student data")

plt.plot(0,0, "o", label="no length takes no time")

plt.xlabel("swingset length, L, (cm)")
plt.ylabel("swingtime, T, (sec)")
plt.xlim(0,1200)
plt.ylim(0,6.5)
plt.legend()
plt.grid()

#plt.show()
plt.savefig("linear_data.pdf")
```

```python

plt.plot(    np.sqrt(l_th),t_th,"-",markerfacecolor="none",label="$T=(2 \pi) \sqrt{L/g}$")

plt.plot(    np.sqrt(length),time,"o",markerfacecolor="none", label="student data")

plt.plot(0,0, "o", label="no length takes no time")


plt.xlabel("swingset length, L, (cm)")
plt.ylabel("swingtime, T, (sec)")
plt.xlim(0,np.sqrt(1200))
plt.ylim(0,6.5)
plt.legend()

plt.grid()

plt.savefig("linearized_data.pdf")
```
