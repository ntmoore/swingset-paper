```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
csv=np.genfromtxt ('./data/fall22-phys180.csv',
                   skip_header=3,
                   delimiter=",")
```

```python
csv_L=np.genfromtxt ('./data/fall22-phys180-large-initial-angle.csv',
                   skip_header=3,
                   delimiter=",")
```

```python
length=csv[:,0]
time=csv[:,1]

length_L=csv_L[:,0]
time_L=csv_L[:,1]

# this is the small-oscillation physics result (eg, Pendulums, Open Stax University Physics, section 16.4)
l_th=np.arange(min(length),max(length))
g=9.81
t_th=(2*np.pi/np.sqrt(g))*(1./10.)*np.sqrt(l_th)

# this is a typical linear fit that students generate with lengths < 100cm
l_th_1=np.arange(min(length),150)
l_th_2=np.arange(150,max(length))
t_linear_1=0.01264*l_th_1+0.8345
t_linear_2=0.01264*l_th_2+0.8345
```

```python

plt.plot(length,time,"o",markerfacecolor="none", label="student data")

plt.plot(length_L,time_L,"x", label="student data, large release angle")
plt.plot(0,0, "o", label="no length takes no time")

plt.plot(l_th_1,t_linear_1,"-",markerfacecolor="none",label="Students\' initial linear model")
#plt.plot(l_th_2,t_linear_2,"--",markerfacecolor="none",label="Students\' initial linear model")
plt.plot(l_th_2,t_linear_2,"--",markerfacecolor="none")

plt.plot(l_th,t_th,"-",markerfacecolor="none",label="$T=(2 \pi) \sqrt{L/g}$")

plt.xlabel("swingset length, L, (cm)")
plt.ylabel("swingtime, T, (sec)")
plt.xlim(0,1200)
plt.ylim(0,6.5)
plt.legend()
plt.grid()

plt.savefig("linear_swingtime_graph.pdf")
```

```python

plt.plot(    np.sqrt(length),time,"o",markerfacecolor="none", label="student data")

plt.plot(    np.sqrt(length_L),time_L,"x", label="student data, large release angle")

plt.plot(0,0, "o", label="no length takes no time")

plt.plot(    np.sqrt(l_th),t_th,"-",markerfacecolor="none",
    label="$T=(2 \pi) \sqrt{L/g} \simeq 0.2 \sqrt{L} ~ sec/ \sqrt{cm} $")
    

plt.xlabel("linearized swingset length, $\sqrt{L}$ ($\sqrt{cm}$)")
plt.ylabel("swingtime, T, (sec)")
plt.xlim(0,np.sqrt(1200))
plt.ylim(0,6.5)
plt.legend()

plt.grid()

plt.savefig("linearized_swingtime_graph.pdf")
```
