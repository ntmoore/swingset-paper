```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
csv=np.genfromtxt ('./data/fall22-phys180.tsv',
                   skip_header=2,
                   delimiter=",")
```

```python
length=csv[:,0]
time=csv[:,1]
```

```python
plt.plot(length,time,"o",markerfacecolor="none")
plt.xlabel("swingset length, L, (cm)")
plt.ylabel("swingtime, T, (sec)")
plt.xlim(0,1200)
plt.ylim(0,6.5)
#plt.legend()
plt.grid()
```

```python
plt.plot(np.sqrt(length),time,"o",markerfacecolor="none")
plt.xlabel("swingset length, L, (cm)")
plt.ylabel("swingtime, T, (sec)")
plt.xlim(0,np.sqrt(1200))
plt.ylim(0,6.5)
#plt.legend()

plt.grid()
```
